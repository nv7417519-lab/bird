```python
import torch
from torchvision import datasets, transforms
import numpy as np
import torch.nn as nn
import matplotlib as plt
import os
```

```python
dir_path = '/Users/bytedance/Desktop/data'
os.makedirs(dir_path, exist_ok = True)
cifar = datasets.CIFAR10(dir_path, train=True, download=True, 
                         transform=transforms.Compose([transforms.ToTensor(), 
                                                       transforms.Normalize(mean=(0.4914, 0.4822, 0.4465),
                                                                            std= (0.2470, 0.2435, 0.2616))]))
cifar_val = datasets.CIFAR10(dir_path, train=False, download=True, 
                             transform=transforms.Compose([transforms.ToTensor(),
                                                           transforms.Normalize(mean=(0.4914, 0.4822, 0.4465),
                                                                                std= (0.2470, 0.2435, 0.2616))]))
```

```python
label_map = {0:0, 2:1}
class_names = ['airplane', 'bird']
cifar2 = [(img, [label_map[label]]) 
          for img, label in cifar 
          if label in [0, 2]]
cifar2_val = [(img, label_map[label]) 
              for img, label in cifar 
              if label in [0, 2]]
```

```python
model = nn.Sequential(
                        nn.Linear(3072, 512),
                        nn.Tanh(),
                        nn.Linear(512, 2),
                        nn.LogSoftmax(dim=1))
loss = nn.NLLLoss()
```

```python
import torch.optim as optim
import torch.nn as nn
train_loader = torch.utils.data.DataLoader(cifar2, shuffle=True, batch_size=64)
epoch = 100
learning_rate = 0.001
loss_fn = nn.NLLLoss()
optimizer = optim.SGD(model.parameters(), lr=learning_rate)
```

```python
for n in range(epoch):
    for img, label in train_loader:
        batch = img.shape[0]
        img_tensor = img.view(batch, -1)
        label_tensor = torch.cat(label)
        out = model(img_tensor)
        loss = loss_fn(out, label_tensor)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        if n%10 == 0: print(f"Epoch: {n}, Loss: {loss:.2f}")
   
```
